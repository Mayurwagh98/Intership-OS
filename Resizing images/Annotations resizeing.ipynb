##Resize

```python
# import os
f = r'/path_of_the_file'
count=0
for file in os.listdir(f):
    f_img = f+"/"+file
    img = Image.open(f_img)
    img = img.resize((1280,720))
    img.save(f_img)
    count = count + 1
    print(count,'-', file,'-', img.size)
```
